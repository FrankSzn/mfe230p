# 1. Convexity

### A. Convex Maximization

Some authors obtain the maximum eigenvalue of a positive semi-definite matrix $A\in\mathbb{R}^{n\times n}$ from the following optimization problem

$$\max_{x} x^\top A x \text{ subject to } \|x\|_2 = 1$$

while others instead write 

$$\max_{x} x^\top A x \text{ subject to } \|x\|_2 \leq 1$$

Despite the fact that the unit sphere $\{ x \mid \|x\|_2 = 1\}$ is a subset (more or less the shell) of the unit ball $\{x \mid \|x\|_2 \leq 1\}$, the two formulations are equivalent (meaning both problems produce the same minimizer) so long as $A \neq \mathbf{0}$. In fact, more generally, maximizing a convex function $f:\mathbb{R}^n \mapsto\mathbb{R}$ over any non-empty closed set $\mathcal{X} \subset \mathbb{R}^n$

$$\max_{x} f(x) \text{ subject to } x \in \mathcal{X}$$

is equivalent to maximizing $f$ over the [convex hull](https://en.wikipedia.org/wiki/Convex_hull) of $\mathcal{X}$

$$\max_{x} f(x) \text{ subject to } x \in \mathbf{conv}(\mathcal{X})$$

when the maxima of $f$ are bounded on $\mathcal{X}$. In fact, this result holds for real [quasi-convex](https://en.wikipedia.org/wiki/Quasiconvex_function) functions, a class of functions that [strictly contains](https://link.springer.com/chapter/10.1007/978-1-4613-0301-5_9) the class of real convex functions. Prove that this claim is indeed true; you are free to choose whether to assume $f$ is convex or quasi-convex.

Utility maximization problems with convex preferences can be modeled as quasi-convex maximization programs, where the feasible region—defined by budget constraints—is a (potentially non-convex) [polytope](https://en.wikipedia.org/wiki/polytope) of some sort. What does the above result suggest about where the optimal basket of goods will lie?

### SOLUTION.

Suppose $p^\star := \max_{x\in\mathcal{X}} f(x)$ where $f$ is convex and $\mathcal{X} \neq \emptyset$. Suppose that all maximizers lie in the set $\mathbf{conv}(\mathcal{X}) \backslash \mathcal{X}$ and let $x_\star$ denote one such maximizer. Then

\begin{align} \exists x,y\in\mathcal{X}: f(x) , &\theta \in (0,1) : x_\star = \theta x + (1-\theta) y \\
\implies p^\star & = f(x_\star) = f(\theta x + (1-\theta) y ) \\
& \leq \theta f(x) + (1-\theta) f(y)\\
& \leq \max\{f(x),f(y)\}
\end{align}

a contradiction of the fact that $p^\star$ is a maximum (when the inequality is strict) or that all maximizers lie in the set $\mathbf{conv}(\mathcal{X}) \backslash \mathcal{X}$ (otherwise). Therefore $\max_{x\in\mathcal{X}} f(x) \equiv \max_{x\in\mathbf{conv}(\mathcal{X})} f(x).$

Given this result, we know that maximizing a quasi-convex utility function over some polytope $\mathcal{P}$ is equivalent to maximizing over any set $\mathcal{Q}$ such that

$$\mathbf{conv}(\mathcal{Q}) = \mathbf{conv}(\mathcal{P})$$

This includes the [boundary](http://bit.ly/2rFPa4h) (i.e. the "outline") of $\mathcal{P}$ as well as the vertices of $\mathcal{P}$ that characterize its shape. In other words, at least one optimal goods basket will lie on a vertex of the budget set. This is in accordance with the classical monotonicity result that states that as long as goods provide utility (i.e. they are goods and not "bads"), the agent will expend their full budget.

### B. Transaction Cost Conventions

Transaction costs play a key role in designing successful trading strategies: If revenues cannot be justified by their costs, then rational investors will not transact. Each broker provides his/her own transaction cost to incentivise certain trading behavior. Consider solving the portfolio optimization

$$\min_x f(x) + \lambda \sum_i T_i(x_i)$$

where $x \in \mathbb{R}^n$ is a vector of portfolio weights such that $x_i$ is the proportion of total funds allocated to the $i^{th}$ asset, $f$ is some measure of risk, and $T_i$ is a function describing transaction cost on $x_i$ weighted by a scalar $\lambda > 0$. In other words, we want to minimize risk while placing a penalty $\lambda$ on transaction costs. For simplicity, let's suppose that $\forall i,j: T_i = T_j$, meaning the same transaction cost protocol applies to all assets.

What conclusion can you draw about the relationship between convex costs and diversification?

### SOLUTION.

**<font color='purple'>Purple.</font>** The purple transaction cost can be modeled by a piecewise constant function

$$T(x_i) = c \cdot I(x_i \neq 0)$$

for some constant $c > 0$, where $I(p)$ is the standard [indicator function](https://en.wikipedia.org/wiki/Indicator_function#Remark_on_notation_and_terminology) which is equal to one if $p$ holds and is equal to zero otherwise. For example,

$$I(x < 0) = \left\{
\begin{array}{ll}
1 & \text{ if } x < 0 \\
0 & \text{ if } x \geq 0
\end{array}
\right.$$

The transaction cost $T_i$, also known as the cardinality function, is non-convex since, for $x_i \neq 0$, the interior of any chord connecting $T_i(0)$ and $T_i(x_i)$ lies below the $T_i$ curve. This transaction cost scheme means the investor only has to pay once and can buy as many shares as he/she likes. This incentivises investors buy fewer securities, but purchase large amounts of each, discouraging diversificaiton. [Scottrade](https://www.scottrade.com) is famous for offering this "flat rate" transaction cost, which works well for casual investors.

**<font color='red'>Red.</font>** The red transaction cost can be modeled by an absolute value function $T_i(x_i) = c \cdot | x_i |$ for some constant $c > 0$. This function is convex since a maximum of convex functions is convex and $|x_i| = \max\{0, x_i\} + \max\{0, -x_i\}$. This means the investor simply pays for the number of shares purchased, irrespective of which asset is traded, encouraging investors to limit the total number of shares (as compared to no costs), but not in a way that would discourage a diversified strategy.

**<font color='green'>Green.</font>** The green transaction cost can be modeled by a piecewise linear function 

$$T_i(x_i) = \min\{ c_1 |x_i|, c_2 |x_i| + b_2 , c_3 |x_i| + b_3 \}$$

for some constants $0 < c_3 < c_2 < c_1$ and $b_3 < b_2 < b_1$. This function is not convex as some chords connecting two points on the total cost curve lie below the total cost curve. For example, the chord connecting $T_i(.1)$ and $T_i(.9)$ lies below $T_i(.5)$. This transaction cost schemes generally encourages investors to purchase more shares as, marginally, shares get cheaper when larger quantities are purhcased. However, like in the flat rate case, this scheme discourages diversification.

**<font color='blue'>Blue.</font>** The blue transaction cost can also be modeled by a piecewise linear function 

$$T_i(x_i) = c\cdot \max\{0, |x_i| - b\}$$

for some constants $b > 0$ and $c > 0$. Since $|\cdot|$ is convex and a $\max$ of convex functions is convex, the transaction cost $T_i$ is convex. The kink at .3 means the investor gets the first few shares for free, then must pay for any additional shares thereafter. This would generally encourage diversification, as investors are paying less on average if they distribute their funds across a wider selection of securities.

The main conclusion we can draw from this simple analysis is that convex transaction costs typically encourage diversification while concave schemes do the opposite. The linear case, which happens to be both convex and concave, neither encourages nor discourages diversification.

# 2. Clustering

### A. $k$-Means Loss is decreasing in $k$

Recall that $k$-means clustering can be expressed as the following mathematical program

$$p^\star(k) = \min_{C_j} \sum_{j=1}^k \sum_{i\in\Omega_j} \|x_i - C_j\|_2^2$$

where $x_i\in\mathbb{R}^n$ is the $i^{th}$ example and $\Omega_j = \left\{i \mid \forall l\neq j: \|x_i - C_j\|_2^2 \leq \|x_i - C_l\|_2^2\right\}$. Show that $p^\star(k+1)\leq p^\star(k)$ for all $k \geq 1$. What are the qualitative implications of this result when applying clustering in practice?

### SOLUTION.

Suppose that $p^\star(k)$ is given and that $\forall j\in \{1,\dots, k\}: C^\star_j = \arg_{C_j} p^\star(k)$. Also, let the objective of the problem $p^\star(k+1)$ be denoted

$$f(C_1,\dots,C_k,C_{k+1}) = \sum_{j=1}^k \sum_{i\in\Omega_j}\|x_i - C_j\|_2^2$$

It should be easy to see that $f(C^\star_1, \dots, C^\star_k, C^\star_j) = p^\star(k)$ for any $j\in\{1,\dots,k\}$. In other words, if we simply set the the first $k$ centroids to what they were in the $p^\star(k)$ problem and set the $(k+1)^{th}$ centroid to any of the other $k$, we obtain the loss $p^\star(k)$. Therefore, minimizing over $C_{k+1}$ should result in a loss that is at most as large as $p^\star(k)$ and moreoever optimizing over all $C_j$ for $j\in\{1,\dots,k+1\}$ will result in an even smaller loss:

$$p^\star(k) \geq \min_{C_{k+1}} f(C^\star_1, \dots, C^\star_k, C_{k+1}) \geq \min_{\{C_1,\dots,C_k,C_{k+1}\}} f(C_1, \dots, C_k, C_{k+1}) = p^\star(k+1)$$

The fact that loss is decreasing in $k$ means that we can always get a higher-fidelity segmentation by increasing the number of cluster centroids to solve for. However, this doesn't necessarily mean that these are the true groupings since, at the limit $k = n$, each observation constitutes its own cluster which is not very informative. The parameter $k$ must be chosen with care using [heuristics](http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_silhouette_analysis.html) that align with the practitioner's hypothesis about how groups form.

### B. Lloyd's Algorithm Implementation

Write a `Python` implementation of Lloyd's algorithm in a function `kmeans` using the template provided in the next cell below. You may not use any external packages to complete this problem aside from `numpy`.

### SOLUTION.

```python
import numpy as np

def kmeans(data, k = 2, total_iter=100):
    """
    data: An (n times p) numpy array with rows representing n observations of dimension p.
    k: An integer specifying the number of desired clusters.
    total_iter: The total number of iterations to perform.
    
    Outputs an array of length n containing cluster assignments.
    """
    
    n = data.shape[0] # size of dataset
    
    # INITIALIZATION

    ind = np.random.randint(
        low=0, high=n, size=k
    ) # draw k observations from dataset to define centroid initialization

    cent = data[ind,:]; cent2 = np.empty(shape = cent.shape) # Cluster centroids
    variance = np.empty(shape = (n, k)) # matrix that will store in-cluster variance
    cluster_assignments = [] # will store data assignment to clusters
    
    # Lloyd's Algorithm

    iteration = 0 # this variable will store iteration number

    while iteration <= total_iter:

        iteration += 1

        for i in range(n):
            for j in range(k):
                # calculate distance between each observation and each cluster center
                variance[i,j] = np.linalg.norm(data[i,:] - cent[j,:])

        # assign each observation to the cluster that it is closest to
        cluster_assignments = np.argmin(variance, axis=1)

        # move centroids to center of each cluster
        for i in range(k):
            cent2[i]= np.mean(data[cluster_assignments == i,:],axis=0)
        cent = cent2

    return cluster_assignments
```

### C. Clusterpath Implementation

Write a `Python` implementation of $\ell_1$-clusterpath

$$\min_{\mu_i,\mu_j} \sum_{i=1}^n \| x_i - \mu_i \|_2^2 + \lambda \sum_{i<j} w_{ij} \|\mu_i - \mu_j\|_1$$

where $x_i$ is a row vector denoting the $i^{th}$ observation and $w_{ij} = \exp\left\{-\|x_i - x_j\|_2^2\right\}$. A template is provided in the next cell below. You may not use any external packages to complete this problem aside from `cvxpy`, `numpy`, and the function `enumerate_centroids` provided in the next cell. A list of `cvxpy` functions can be accessed [here](http://www.cvxpy.org/en/latest/tutorial/functions/).

Intuitively, what role do the fusion penalty weights $w_{ij}$ play in clusterpath? What do you expect to occur when $\forall i,j: w_{ij} = 1$? 

### SOLUTION.

```python
import cvxpy as cvx

def enumerate_centroids(centroids):
    """Given a n x p matrix of centroids outputted from cvxpy,
    where each of the n rows represents a centroid, will return
    an n x 1 array enumerating centroids."""

    unique = {} # stores unique centroids and numbers
    enumeration = [] # to store cluster number
    i = 0 # cluster number

    for row in np.round(centroids, 3):

        key = tuple(row)
        value = unique.get(key)

        if value is not None:
            enumeration.append(value)
        else:
            unique[key] = i # add cluster to dictionary
            enumeration.append(i)
            i += 1 # increment counter by one

    return np.array(enumeration)

def clusterpath(data, lamb=1, gamma=1):
    """
    data: An (n times p) numpy array with rows representing n 
    observations of dimension p.
    lamb: A positive number representing lambda in the clusterpath
    problem, which controls how harshly to penalize the fusion
    penalties.
    gamma: The bandwidth parameter of the exponentially-decaying weights.
    
    Outputs an array of length n containing cluster assignments.
    """

    n, p = data.shape

    MU = cvx.Variable(n, p) # centroid assigned to each point (will be fused)
    fusion_penalties = [] # will store |mu_i - mu_j|_1 penalties

    for i in range(1, n):
        for j in range(i):

            weight_ij = np.exp( # squared distance between datum
                -gamma*np.linalg.norm(data[i, :] - data[j, :], 2)**2
            )

            fusion_penalties.append( # fusion penalty weighted by distance
                weight_ij * cvx.norm(MU[i, :] - MU[j, :], 1)
            )

    vec_fp = cvx.vstack(*fusion_penalties) # vector of fusion penalties

    objective = cvx.Minimize(
        cvx.square(cvx.norm(data - MU, 'fro')) + lamb*cvx.sum_entries(vec_fp),
    )

    pstar = cvx.Problem(objective).solve()
    
    cluster_assignments = enumerate_centroids(MU.value)
    
    return cluster_assignments
```

Having $w_{ij}$ be decreasing in $\|x_i - x_j\|$ means that the fusion penalty on centroids is larger if observations are close to each other and decays quickly if they are farther apart. Intuitively, this means that fusing two far-apart cluster centroids will not decrease the loss substantially, so clusterpath will instead focus on fusing clusters that are adjacent, per say. Focusing on adjancency means capturing local information, like forming clusters that may not necessarily be spherically-shaped. When $w_{ij} = 1$, the fusion penalties are insensitive to distance between observations and therefore will try to fuse the centroids of two points that may be arbitrarily far apart.

### SOLUTION.

Clusterpath is called so because it generates a path for each centroid $\{\mu_i(\lambda)\mid \lambda > 0\}$ that contains no re-splits after fusions. This means, unlike $k$-means, if we increase $\lambda$ to obtain one more cluster, it will not significantly displace the existing cluster assignments. Instead, it will likely take a few observations that "don't fit in well" with existing clusters and form a new cluster therewith. On the other hand, each time $k$ is changed, $k$-means starts from scratch and therfore can have significantly different centroids and partitions depending on $k$. This follows directly from the fact that the objective of $k$-means encourages larger, bulkier clusters that cover more space, especially centered around dense sets of observations.

Also, as noted earlier, clusterpath with exponentially-decaying weights has an advantage in its ability to capture local information, where the tradeoff between local and global is controlled by $\gamma$. We will later discuss spectral $k$-means, which operates under the $k$-means framework but is also locally-sensitive.

### G. $\ell_1$ distance metric

Suppose for both $k$-means and clusterpath, we replaced the $\ell_2$ norm for computing distance to clusters with the $\ell_1$ norm. Specifically, consider obtaining clusters by solving 

$$\min_{C_j} \sum_{j=1}^k \left\{ \frac{1}{\left|\Omega_j\right|} \sum_{i\in\Omega_j} \|x_i - C_j\|_1 \right\}$$

instead of $k$-means or 

$$\min_{\mu_i,\mu_j} \sum_{i=1}^n \| x_i - \mu_i \|_1 + \lambda \sum_{i<j} w_{ij} \|\mu_i - \mu_j\|_1$$

instead of clusterpath. Explain qualitatively how this would affect the way clusters were assigned in the previous excercise.

### SOLUTION.

As discussed in lecture, the $\ell_1$ norm is more robust to outliers, meaning that in a space defined by the $\ell_1$ distance metric, outlying observations are significantly closer to inlying ones. Contrapositively, this means observations that may have been very close in $\ell_2$ distance are now spread a bit farther form each other. Loosely, in an $\ell_1$ metric space, natural distance is often (but not always) de-emphasized.

This is a particularly attractive benefit when working with financial data—such as returns, which are frequently leptokurtic—and would mitigate the issue of having outliers form their own clusters. Furthermore, computationally, the $\ell_1$-norm versions of clusterpath is defined by a linear program (can you see why?) meaning the solution time is potentially much faster than the original formulation.
